```python
files = glob.glob("states*.csv")
states_list = []
for filename in files:
    data = pd.read_csv(filename)
    states_list.append(data)

us_census = pd.concat(states_list)
```

## Recovering data

```python
us_census["Income"] = (us_census["Income"].replace('[\$,]', '', regex=True))
us_census["Income"] = (us_census["Income"].replace('[\,]', '', regex=True))


us_census["Income"] = pd.to_numeric(us_census["Income"])

us_census["Income"]
```

```python
us_census["GenderPop"].head()

us_census[["Men Count", "Women Count"]] = us_census["GenderPop"].str.split("_", expand=True)

us_census["Men Count"] = pd.to_numeric(us_census["Men Count"].str.replace("M", "", regex=False))
us_census["Women Count"] = pd.to_numeric(us_census["Women Count"].str.replace("F", "", regex=False))

us_census['Women Count'] = us_census['Women Count'].fillna(us_census['TotalPop'] - us_census['Men Count']).astype(int)
```
