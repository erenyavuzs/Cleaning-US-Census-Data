```python
files = glob.glob("states*.csv")
states_list = []
for filename in files:
    data = pd.read_csv(filename)
    states_list.append(data)

us_census = pd.concat(states_list)
```

## Recovering data

```python
us_census["Income"] = (us_census["Income"].replace('[\$,]', '', regex=True))
us_census["Income"] = (us_census["Income"].replace('[\,]', '', regex=True))


us_census["Income"] = pd.to_numeric(us_census["Income"])

us_census["Income"]
```

```python
us_census["GenderPop"].head()

us_census[["Men_Count", "Women_Count"]] = us_census["GenderPop"].str.split("_", expand=True)

us_census["Men_Count"] = pd.to_numeric(us_census["Men_Count"].str.replace("M", "", regex=False))
us_census["Women_Count"] = pd.to_numeric(us_census["Women_Count"].str.replace("F", "", regex=False))

us_census['Women_Count'] = us_census['Women_Count'].fillna(us_census['TotalPop'] - us_census['Men_Count']).astype(int)
```

```python
race_columns = ["Hispanic", "White", "Black", "Native", "Asian", "Pacific"]

for col in race_columns:
    us_census[col] = pd.to_numeric(us_census[col].str.replace("%", "", regex=False))
```

```python
us_census.rename(columns={
    "Hispanic": "Hispanic_Pct",
    "White": "White_Pct",
    "Black": "Black_Pct",
    "Native": "Native_Pct",
    "Asian": "Asian_Pct",
    "Pacific": "Pacific_Pct"
}, inplace=True)
```
